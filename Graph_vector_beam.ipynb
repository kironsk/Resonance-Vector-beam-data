```python
dh_1=pd.read_csv("radial sem corretor 18-09-23.csv")
dh_2=pd.read_csv("radial com corretor 18-09-23.csv")
dh_3=pd.read_csv("azimutal sem corretor 18-09-23.csv")
dh_4=pd.read_csv("azimutal com corretor 18-09-23.csv")
```

```python
dh1 = dh_1.loc[(dh_1.iloc[:,3]>-0.0260) & (dh_1.iloc[:,3]<0.0150)]
dh2 = dh_2.loc[(dh_2.iloc[:,3]>-0.0105) & (dh_2.iloc[:,3]<0.026)]
dh3 = dh_3.loc[(dh_3.iloc[:,3]>-0.021) & (dh_3.iloc[:,3]<0.017)]
dh4 = dh_4.loc[(dh_4.iloc[:,3]>-0.0275) & (dh_4.iloc[:,3]<0.011)]
dh5 = dh_1.loc[(dh_1.iloc[:,3]>0.008) & (dh_1.iloc[:,3]<0.0130)]
```

```python
x_radial_sem_ = dh1.iloc[:,3]
y_radial_sem_ = dh1.iloc[:,4]

x_radial_com_ = dh2.iloc[:,3]
y_radial_com_ = dh2.iloc[:,4]

x_azimutal_sem_ = dh3.iloc[:,3]
y_azimutal_sem_ = dh3.iloc[:,4]

x_azimutal_com_ = dh4.iloc[:,3]
y_azimutal_com_ = dh4.iloc[:,4]

x_corte_ = dh5.iloc[:,3]
y_corte_ = dh5.iloc[:,4]


```

```python
x_radial_sem = x_radial_sem_.rolling(4).median()
y_radial_sem = y_radial_sem_.rolling(4).median()

x_radial_com = x_radial_com_.rolling(4).median()
y_radial_com = y_radial_com_.rolling(4).median()

x_azimutal_sem = x_azimutal_sem_.rolling(4).median()
y_azimutal_sem = y_azimutal_sem_.rolling(4).median()

x_azimutal_com = x_azimutal_com_.rolling(4).median()
y_azimutal_com = y_azimutal_com_.rolling(4).median()

x_corte = x_corte_.rolling(4).median()
y_corte = y_corte_.rolling(4).median()
```

```python
fig, ax = plt.subplots(figsize=(12,8),nrows= 2 , ncols=2)
ax[0,0].plot(x_radial_sem, y_radial_sem)
ax[0,0].set_ylabel("Intensity (a.u.)", fontsize =15)



ax[0,1].plot(x_radial_com, y_radial_com)



ax[1,0].plot(x_azimutal_sem, y_azimutal_sem)

ax[1,1].plot(x_azimutal_com, y_azimutal_com)
plt.savefig("h.svg")
plt.show()
```
